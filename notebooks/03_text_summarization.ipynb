```python
from transformers import pipeline
import os
```

```python
#LOAD TRANSCRIPT FROM FILE
transcript_path = os.path.join("..", "data", "transcripts", "video_transcript.txt")

with open(transcript_path, "r", encoding="utf-8") as f:
    text = f.read()

print("Transcript length (words):", len(text.split()))
```

```python
#LOAD SUMMARIZATION MODEL
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)
```

```python
summary = summarizer(
    text,
    max_length=40,
    min_length=10,
    do_sample=False
)
#Note: Input transcript is short; summarization quality improves with longer input text.
```

```python
#VIEW SUMMARY
print(summary[0]["summary_text"])
```

```python
#SAVE SUMMARY TO FILE
summary_path = os.path.join("..", "data", "summaries", "video_summary.txt")

with open(summary_path, "w", encoding="utf-8") as f:
    f.write(summary[0]["summary_text"])

print("Summary saved to:", summary_path)
```
