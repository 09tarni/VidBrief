```python
from IPython.display import Audio
import os

audio_path = os.path.join("..", "data", "audio", "speech_audio.wav")
print("Exists:", os.path.exists(audio_path))
print("Size:", os.path.getsize(audio_path))

Audio(audio_path)
```

```python
import whisper

model = whisper.load_model("base")
result = model.transcribe(audio_path)
print(result["text"])
```

```python
output_path = os.path.join("..", "data", "transcripts", "video_transcript.txt")

with open(output_path, "w", encoding="utf-8") as f:
    f.write(result["text"])

print("Transcript saved to:", output_path)
```
