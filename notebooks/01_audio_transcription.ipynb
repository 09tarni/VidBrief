```python
import os

audio_path = os.path.join("..", "data", "librispeech_sample", "sample.flac")
print("File exists:", os.path.exists(audio_path))
```

```python
import whisper

model = whisper.load_model("base")
```

```python
#TRANSCRIBE AUDIO
```

```python
result = model.transcribe(audio_path)
```

```python
print(result["text"])
```

```python
output_path = os.path.join("..", "data", "transcripts", "sample_transcript.txt")

with open(output_path, "w", encoding="utf-8") as f:
    f.write(result["text"])

print("Saved to:", output_path)
```
